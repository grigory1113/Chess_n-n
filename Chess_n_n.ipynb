```python
class Available_moves:
    def __init__(self, x, y, size=size):
      self.x = x
      self.y = y
      self.size = size

    def direct(self, position):
        moves = []
        for i in range(min(self.x+1, self.size-1), self.size):
            if not issubclass(type(position[i][self.y]), Piece):
                moves.append((self.x, self.y, i, self.y))
            else:
                if position[self.x][self.y].color != position[i][self.y].color:
                    moves.append((self.x, self.y, i, self.y))
                break
        for i in range(max(0, self.x-1), -1, -1):
            if not issubclass(type(position[i][self.y]), Piece):
                moves.append((self.x, self.y, i, self.y))
            else:
                if position[self.x][self.y].color != position[i][self.y].color:
                    moves.append((self.x, self.y, i, self.y))
                break
        for i in range(min(self.y+1, self.size-1), self.size):
            if not issubclass(type(position[self.x][i]), Piece):
                moves.append((self.x, self.y, self.x, i))
            else:
                if position[self.x][self.y].color != position[self.x][i].color:
                    moves.append((self.x, self.y, self.x, i))
                break
        for i in range(max(0, self.y-1), -1, -1):
            if not issubclass(type(position[self.x][i]), Piece):
                moves.append((self.x, self.y, self.x, i))
            else:
                if position[self.x][self.y].color != position[self.x][i].color:
                    moves.append((self.x, self.y, self.x, i))
                break

        return moves

    def diagonal(self, position):
        moves = []
        for i in range(1, self.size):
          if self.x+i in range(self.size) and self.y+i in range(self.size):
            if not issubclass(type(position[self.x+i][self.y+i]), Piece):
                moves.append((self.x, self.y, self.x+i, self.y+i))
            else:
                if position[self.x][self.y].color != position[self.x+i][self.y+i].color:
                    moves.append((self.x, self.y, self.x+i, self.y+i))
                break
          else:
            break
        for i in range(1, self.size):
          if self.x-i in range(self.size) and self.y+i in range(self.size):
            if not issubclass(type(position[self.x-i][self.y+i]), Piece):
                moves.append((self.x, self.y, self.x-i, self.y+i))
            else:
                if position[self.x][self.y].color != position[self.x-i][self.y+i].color:
                    moves.append((self.x, self.y, self.x-i, self.y+i))
                break
          else:
            break
        for i in range(1, self.size):
          if self.x+i in range(self.size) and self.y-i in range(self.size):
            if not issubclass(type(position[self.x+i][self.y-i]), Piece):
                moves.append((self.x, self.y, self.x+i, self.y-i))
            else:
                if position[self.x][self.y].color != position[self.x+i][self.y-i].color:
                    moves.append((self.x, self.y, self.x+i, self.y-i))
                break
          else:
            break
        for i in range(1, self.size):
          if self.x-i in range(self.size) and self.y-i in range(self.size):
            if not issubclass(type(position[self.x-i][self.y-i]), Piece):
                moves.append((self.x, self.y, self.x-i, self.y-i))
            else:
                if position[self.x][self.y].color != position[self.x-i][self.y-i].color:
                    moves.append((self.x, self.y, self.x-i, self.y-i))
                break
          else:
            break

        return moves
```

```python
class Piece:
    def __init__(self, x, y, color, size=size):
        self.x = x
        self.y = y
        self.color = color
        self.size = size

    def __repr__(self):
        return self.img[self.color]


class King(Piece):
    def __init__(self, x, y, color, size=size, check=0):
        super().__init__(x, y, color, size)
        self.name = 'K'
        self.img = ('♔', '♚')
        self.check = check


    def get_moves(self, position):
        moves = []
        a = [-1, 0, 1]
        for i in a:
          if self.x+i in range(self.size):
            for j in a:
              if self.y+j in range(self.size):
                if (i, j) != (0, 0):
                  if not issubclass(type(position[self.x+i][self.y+j]), Piece):
                    moves.append((self.x, self.y, self.x+i, self.y+j))
                  else:
                    if position[self.x][self.y].color != position[self.x+i][self.y+j].color:
                      moves.append((self.x, self.y, self.x+i, self.y+j))
        return moves


class Queen(Piece):
    def __init__(self, x, y, color, size=size):
        super().__init__(x, y, color, size)
        self.name = 'Q'
        self.img = ('♕', '♛')

    def get_moves(self, position):
        a = Available_moves(self.x, self.y)
        moves1 = a.direct(position)
        moves2 = a.diagonal(position)
        return moves1 + moves2


class Rook(Piece):
    def __init__(self, x, y, color, size=size):
        super().__init__(x, y, color, size)
        self.name = 'R'
        self.img = ('♖', '♜')

    def get_moves(self, position):
        a = Available_moves(self.x, self.y)
        moves = a.direct(position)
        return moves


class Bishop(Piece):
    def __init__(self, x, y, color, size=size):
        super().__init__(x, y, color, size)
        self.name = 'B'
        self.img = ('♗', '♝')

    def get_moves(self, position):
        a = Available_moves(self.x, self.y)
        moves = a.diagonal(position)
        return moves


class Knight(Piece):
    def __init__(self, x, y, color, size=size):
        super().__init__(x, y, color, size)
        self.name = 'N'
        self.img = ('♘', '♞')

    def get_moves(self, position):
        moves = []
        a = [-1, 1]
        b = [-2, 2]
        for i in a:
          for j in b:
            if self.x+i in range(self.size) and self.y+j in range(self.size):
               if not issubclass(type(position[self.x+i][self.y+j]), Piece):
                  moves.append((self.x, self.y, self.x+i, self.y+j))
               else:
                  if position[self.x][self.y].color != position[self.x+i][self.y+j].color:
                      moves.append((self.x, self.y, self.x+i, self.y+j))
        for i in b:
          for j in a:
            if self.x+i in range(self.size) and self.y+j in range(self.size):
               if not issubclass(type(position[self.x+i][self.y+j]), Piece):
                  moves.append((self.x, self.y, self.x+i, self.y+j))
               else:
                  if position[self.x][self.y].color != position[self.x+i][self.y+j].color:
                      moves.append((self.x, self.y, self.x+i, self.y+j))
        return moves


class Pawn(Piece):
    def __init__(self, x, y, color, size=size):
        super().__init__(x, y, color, size)
        self.name = ''
        self.img = ('♙', '♟')
        self.last_move = (0, 0, 0, 0)

    def get_moves(self, position):
        moves = []
        a = [-1, 1]
        if self.color == 0:
          if self.x < self.size - 2:
            if not issubclass(type(position[self.x+1][self.y]), Piece):
              moves.append((self.x, self.y, self.x+1, self.y))
              if self.x == 1 and not issubclass(type(position[self.x+2][self.y]), Piece):
                moves.append((self.x, self.y, self.x+2, self.y))
          elif self.x == self.size - 2 and not issubclass(type(position[self.x+1][self.y]), Piece):
            turn_pieces = ['Q', 'R', 'B', 'N']
            for piece in turn_pieces:
              moves.append((self.x, self.y, (self.x+1, piece), self.y))
          for i in a:
            if self.y+i in range(self.size):
              if issubclass(type(position[self.x+1][self.y+i]), Piece):
                if self.x < self.size - 2:
                  if position[self.x][self.y].color != position[self.x+1][self.y+i].color:
                    moves.append((self.x, self.y, self.x+1, self.y+i))
                elif self.x == self.size - 2:
                  turn_pieces = ['Q', 'R', 'B', 'N']
                  for piece in turn_pieces:
                    moves.append((self.x, self.y, (self.x+1, piece), self.y+i))
              elif isinstance(position[self.x][self.y+i], Pawn):
                if position[self.x][self.y].color != position[self.x][self.y+i].color and position[self.x][self.y+i].last_move[2]-position[self.x][self.y+i].last_move[0] == -2:
                  moves.append((self.x, self.y, [self.x+1, ''], self.y+i))
        else:
          if not issubclass(type(position[self.x-1][self.y]), Piece):
            if self.x > 1:
              moves.append((self.x, self.y, self.x-1, self.y))
              if self.x == self.size - 2 and not issubclass(type(position[self.x-2][self.y]), Piece):
                moves.append((self.x, self.y, self.x-2, self.y))
            elif self.x == 1 and not issubclass(type(position[self.x-1][self.y]), Piece):
              turn_pieces = ['Q', 'R', 'B', 'N']
              for piece in turn_pieces:
                moves.append((self.x, self.y, (self.x-1, piece), self.y))
          for i in a:
            if self.y+i in range(self.size):
              if issubclass(type(position[self.x-1][self.y+i]), Piece):
                if self.x > 1:
                  if position[self.x][self.y].color != position[self.x-1][self.y+i].color:
                    moves.append((self.x, self.y, self.x-1, self.y+i))
                elif self.x == 1:
                  turn_pieces = ['Q', 'R', 'B', 'N']
                  for piece in turn_pieces:
                    moves.append((self.x, self.y, (self.x-1, piece), self.y+i))
              elif isinstance(position[self.x][self.y+i], Pawn):
                if position[self.x][self.y].color != position[self.x][self.y+i].color and position[self.x][self.y+i].last_move[2]-position[self.x][self.y+i].last_move[0] == 2:
                  moves.append((self.x, self.y, [self.x-1, ''], self.y+i))

        return moves

    def turn_into(self, piece = 'Q'):
      if piece == 'N':
        self = Knight(self.x, self.y, self.color)
      elif piece == 'R':
        self = Rook(self.x, self.y, self.color)
      elif piece == 'Q':
        self = Queen(self.x, self.y, self.color)
      elif piece == 'B':
        self = Bishop(self.x, self.y, self.color)
      return self
```

```python
class Position:
    def __init__(self, position = start_position, size = size, turn = 0):
        self.position = position
        self.turn = turn
        self.size = size
        self.hist = []

    def __repr__(self):
        b = [list(map(str, self.position[i])) for i in range(self.size)]
        res = ''
        for t in range(self.size):
            res += ''.join(b[t]) + '\n'
        return res


    def get_moves(self, x, y):
        moves = self.position[x][y].get_moves(self.position)
        return moves

    def make_xy_move(self, x, y):
        lst = self.position[x][y].get_moves(self.position)
        print(lst)
        a = int(input())
        self.position[x][y].x, self.position[x][y].y = lst[a][2], lst[a][3]
        self.position[lst[a][2]][lst[a][3]] = self.position[x][y]
        self.position[x][y] = '— '
        self.turn = (self.turn + 1) % 2
        return self

    def make_move(self, mode = 'game1'):

        def procedure(flag = 1):
          if self.is_mate():
            print('Mate!')
            return 1
          elif self.is_pat():
            print('Pat!')
            return 1
          elif self.is_draw():
            print('Draw!')
            return 1
          lst = self.get_all_moves()
          available_moves = []
          j = Notation()
          for i, item in enumerate(lst):
            name = self.position[item[0]][item[1]].name
            l1 = j.transform(item[0], item[1])
            if isinstance(item[2], tuple) or isinstance(item[2], list):
              l2 = j.transform(item[2][0], item[3])
              move = f'{name}{l1}-{l2}{item[2][1]}'
              available_moves.append((i, move))
            else:
              l2 = j.transform(item[2], item[3])
              move = f'{name}{l1}-{l2}'
              available_moves.append((i, move))
          def procedure1():
            if isinstance(self.position[lst[a][0]][lst[a][1]], Pawn):
              move_to_write = (lst[a][0], lst[a][1], lst[a][2], lst[a][3])
              if isinstance(move_to_write[2], int):
                self.position[lst[a][0]][lst[a][1]].last_move = move_to_write
              else:
                self.position[lst[a][0]][lst[a][1]].last_move = (lst[a][0], lst[a][1], lst[a][2][0], lst[a][3])
              if isinstance(lst[a][2], tuple):
                self.position[lst[a][0]][lst[a][1]].x, self.position[lst[a][0]][lst[a][1]].y = lst[a][2][0], lst[a][3]
                self.position[lst[a][2][0]][lst[a][3]] = self.position[lst[a][0]][lst[a][1]]
                self.position[lst[a][2][0]][lst[a][3]] = self.position[lst[a][2][0]][lst[a][3]].turn_into(lst[a][2][1])
                self.position[lst[a][0]][lst[a][1]] = '— '
              elif isinstance(lst[a][2], list):
                self.position[lst[a][0]][lst[a][1]].x, self.position[lst[a][0]][lst[a][1]].y = lst[a][2][0], lst[a][3]
                self.position[lst[a][2][0]][lst[a][3]] = self.position[lst[a][0]][lst[a][1]]
                self.position[lst[a][0]][lst[a][1]] = '— '
                delta = 0
                if self.position[lst[a][2][0]][lst[a][3]].color == 0:
                  delta = -1
                else:
                  delta = 1
                self.position[lst[a][2][0] + delta][lst[a][3]] = '— '
              else:
                self.position[lst[a][0]][lst[a][1]].x, self.position[lst[a][0]][lst[a][1]].y = lst[a][2], lst[a][3]
                self.position[lst[a][2]][lst[a][3]] = self.position[lst[a][0]][lst[a][1]]
                self.position[lst[a][0]][lst[a][1]] = '— '
            else:
              self.position[lst[a][0]][lst[a][1]].x, self.position[lst[a][0]][lst[a][1]].y = lst[a][2], lst[a][3]
              self.position[lst[a][2]][lst[a][3]] = self.position[lst[a][0]][lst[a][1]]
              self.position[lst[a][0]][lst[a][1]] = '— '
            self.turn = (self.turn + 1) % 2
            copy_position = []
            for item1 in self.position:
              g = []
              for item2 in item1:
                if isinstance(item2, str):
                  g.append('— ')
                else:
                  g.append((item2.name, item2.color, self.turn))
              copy_position.append(g)
            self.hist.append(copy_position)
            if self.is_mate():
              print('Mate!')
              return 1
            elif self.is_pat():
              print('Pat!')
              return 1
            elif self.is_draw():
              print('Draw!')
              return 1
          if flag == 1:
            print(available_moves)
            a = int(input())
            procedure1()
          elif flag == 0:
            available_moves1 = [available_moves[i][1] for i in range(len(available_moves))]
            b = input()
            if b in available_moves1:
              a = available_moves1.index(b)
              procedure1()
            else:
              print('Impossible move')


        if mode == 'move':
          procedure()
          return self
        elif mode == 'game1':
          while True:
            p = procedure()
            print(self)
            if p == 1:
              break
        elif mode == 'game2':
          while True:
            p = procedure(flag = 0)
            print(self)
            if p == 1:
              break

    def get_all_moves(self):
        all_moves = []
        for i in range(self.size):
          for j in range(self.size):
            if issubclass(type(self.position[i][j]), Piece):
              if self.position[i][j].color == self.turn:
                all_moves.extend(self.get_moves(i, j))
        possible_moves = []
        for item in all_moves:
          p = Analyzer(self)
          if p.try_move(item[0], item[1], item[2], item[3]).is_check() != 1:
            possible_moves.append(item)
        return possible_moves

    def is_mate(self):
      lst = self.get_all_moves()
      if len(lst) == 0 and self.check_now() == 1:
        return 1

    def is_pat(self):
      lst = self.get_all_moves()
      if len(lst) == 0 and self.check_now() != 1:
        return 1


    def is_draw(self):
      black_pieces = []
      white_pieces = []
      for i in range(self.size):
        for j in range(self.size):
          if issubclass(type(self.position[i][j]), Piece):
            if self.position[i][j].color == 0:
              white_pieces.append(self.position[i][j].name)
            elif self.position[i][j].color == 1:
              black_pieces.append(self.position[i][j].name)
      if (len(black_pieces) == 2 and 'N' in black_pieces and
          len(white_pieces) == 1) or (len(black_pieces) == 2 and 'B' in black_pieces
              and len(white_pieces) == 1) or (len(white_pieces) == 2
                  and 'B' in white_pieces and len(black_pieces) == 1) or (len(white_pieces) == 2
                      and 'N' in white_pieces and len(black_pieces) == 1) or (len(black_pieces) == 1 and len(white_pieces) == 1):
                      return 1
      if len(self.hist) >= 6:
        for i in range(len(self.hist)):
          for j in range(i+1, len(self.hist)):
            for z in range(j+1, len(self.hist)):
              if self.hist[i] == self.hist[j] == self.hist[z]:
                return 1

    def is_check(self):
        x = None
        y = None
        for i in range(self.size):
          for j in range(self.size):
            if isinstance(self.position[i][j], King) and self.position[i][j].color != self.turn:
              x, y = i, j
              break
        for i in range(self.size):
          for j in range(self.size):
            if issubclass(type(self.position[i][j]), Piece):
              if self.position[i][j].color == self.turn:
                c = self.get_moves(i, j)
                w = []
                for k in range(len(c)):
                  if isinstance(c[k][2], int):
                    w.append((c[k][2], c[k][3]))
                  else:
                    w.append((c[k][2][0], c[k][3]))
                if (x, y) in w:
                  return 1
        return 0

    def check_now(self):
        x = 0
        y = 0
        for i in range(self.size):
          for j in range(self.size):
            if isinstance(self.position[i][j], King) and self.position[i][j].color == self.turn:
              x, y = i, j
              break
        for i in range(self.size):
          for j in range(self.size):
            if issubclass(type(self.position[i][j]), Piece):
              if self.position[i][j].color != self.turn:
                c = self.get_moves(i, j)
                w = []
                for k in range(len(c)):
                  if isinstance(c[k][2], int):
                    w.append((c[k][2], c[k][3]))
                  else:
                    w.append((c[k][2][0], c[k][3]))
                if (x, y) in w:
                  return 1
        return 0


    def estimate(self):
        s = 0
        for i in range(self.size):
            for j in range(self.size):
                if isinstance(self.position[i][j], Pawn):
                    if self.position[i][j].color == 0:
                        s += 1
                    else:
                        s += -1
                elif isinstance(self.position[i][j], Knight):
                    if self.position[i][j].color == 0:
                        s += 3
                    else:
                        s += -3
                elif isinstance(self.position[i][j], Bishop):
                    if self.position[i][j].color == 0:
                        s += 3
                    else:
                        s += -3
                elif isinstance(self.position[i][j], Rook):
                    if self.position[i][j].color == 0:
                        s += 5
                    else:
                        s += -5
                elif isinstance(self.position[i][j], Queen):
                    if self.position[i][j].color == 0:
                        s += 9
                    else:
                        s += -9
        if self.is_pat() or self.is_draw():
            s = 0
        elif self.is_mate():
            if self.turn == 0:
                s = -1000
            else:
                s = 1000
        return s
```

```python
class Analyzer:
    def __init__(self, obj):
        self.obj = copy.deepcopy(obj)

    def try_move(self, x1, y1, x2, y2):
        if isinstance(x2, tuple):
          self.obj.position[x1][y1].x, self.obj.position[x1][y1].y = x2[0], y2
          self.obj.position[x2[0]][y2] = self.obj.position[x1][y1]
          self.obj.position[x1][y1] = '— '
          self.obj.position[x2[0]][y2] = self.obj.position[x2[0]][y2].turn_into(x2[1])
        elif isinstance(x2, list):
          self.obj.position[x1][y1].x, self.obj.position[x1][y1].y = x2[0], y2
          self.obj.position[x2[0]][y2] = self.obj.position[x1][y1]
          self.obj.position[x1][y1] = '— '
          delta = 0
          if self.obj.position[x2[0]][y2].color == 0:
            delta == -1
          else:
            delta == 1
          self.obj.position[x2[0] + delta][y2] = '— '
        else:
          self.obj.position[x1][y1].x, self.obj.position[x1][y1].y = x2, y2
          self.obj.position[x2][y2] = self.obj.position[x1][y1]
          self.obj.position[x1][y1] = '— '
        self.obj.turn = (self.obj.turn + 1) % 2
        copy_position = []
        for item1 in self.obj.position:
          g = []
          for item2 in item1:
            if isinstance(item2, str):
              g.append('— ')
            else:
              g.append((item2.name, item2.color, self.obj.turn))
          copy_position.append(g)
        self.obj.hist.append(copy_position)
        return self.obj
```

```python
class Engine:
  def __init__(self, obj):
    self.obj = obj

  def best_move(self):
    estimates = []
    all_moves = self.obj.get_all_moves()
    p = Analyzer(self.obj)
    for move in all_moves:
      pos = p.try_move(move[0], move[1], move[2], move[3])
      estimates.append((move, pos, pos.estimate()))
      p = Analyzer(self.obj)
    est = [estimates[j][2] for j in range(len(estimates))]
    if self.obj.turn == 0:
      m = max(est)
      i = est.index(m)
      self.obj = estimates[i][1]
    else:
      m = min(est)
      i = est.index(m)
      self.obj = estimates[i][1]
    return self.obj

  def engine_game(self, max_number = 100):
    i = 0
    while True:
      q = self.best_move()
      self.obj = q
      print(self.obj)
      i += 1
      if i >= max_number:
        return
      if self.obj.is_mate():
        print('Mate!')
        return
      elif self.obj.is_pat() or self.obj.is_draw():
        print('Draw!')
        return
```
